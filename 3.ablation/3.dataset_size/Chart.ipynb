```python
embedding_score_dict = {
    "e5large": 0.8525,
    "bge-m3": 0.8515,
    "w2v (Ekstra)": 0.8476,
    "bert (Ekstra)": 0.8427,
    "contrastive (Ekstra)": 0.8507,
    "xlm (Ekstra)": 0.8458,
    "image (Ekstra)": 0.843,
    
}
```

```python
with plt.style.context(['science']):
    fig, ax = plt.subplots()
    pparam = dict(xlabel='Embedding Types', ylabel="ROC AUC")
    ax.set(**pparam)
    
    bars = ax.bar(embedding_score_dict.keys(), embedding_score_dict.values())
        
    # Adding bar labels
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.4f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', size=8)
    
    ax.set_ylim(0.84, 0.855)
    ax.set_xticklabels(list(embedding_score_dict.keys()), rotation=60, ha='right')
    plt.show()
```
